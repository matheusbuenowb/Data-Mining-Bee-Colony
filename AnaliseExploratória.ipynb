#### (1) `Box Plot`
* Resumo das propriedades de um conjunto de valores;
* Ex: mínimo, primeiro quartil, mediana, terceiro quartil e máximo

```python
df.columns
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(data = df['hive temp'])
```

```python
plt.figure(figsize=(18, 6))  # aumentei a largura para acomodar mais gráficos

def create_boxplot(feature):
    sns.boxplot(data=df[[feature]])
    plt.title(feature, fontsize=10)

features = ['hive temp', 'hive humidity', 'hive pressure',
            'weather temp', 'weather humidity', 'weather pressure']

for i, feature in enumerate(features):
    plt.subplot(1, 6, i+1)
    create_boxplot(feature)

plt.tight_layout()
plt.show()
```

#### Tratando outliers

```python
df.shape
```

```python
# quantas vezes o outlier se repete
print(df['hive temp'].loc[df['hive temp']>= 45].size)
print(df['hive humidity'].loc[df['hive humidity']>= 83].size)
print(df['hive pressure'].loc[df['hive pressure']>= 1015].size)
```

```python
# Antes da substituição
print(df['hive temp'].mean())
print(df['hive humidity'].mean())
print(df['hive pressure'].mean())
```

```python
# Calcular a média sem o outlier
media_hive_temp = df['hive temp'].loc[df['hive temp'] < 45].mean()
media_hive_humidity = df['hive humidity'].loc[df['hive humidity'] < 83].mean()
media_hive_pressure = df['hive pressure'].loc[df['hive pressure'] < 1015].mean()
```

```python
print(media_hive_temp)
print(media_hive_humidity)
print(media_hive_pressure)
```

```python
# Substui o outlier pela média
# linha/coluna
df.loc[df['hive temp'] >= 45,'hive temp'] = media_hive_temp
df.loc[df['hive humidity'] >= 83,'hive humidity'] = media_hive_humidity
df.loc[df['hive pressure'] >= 1015,'hive pressure'] = media_hive_pressure
```

```python
#  Após a substituição
print(df['hive temp'].mean())
print(df['hive humidity'].mean())
print(df['hive pressure'].mean())
```

```python
plt.figure(figsize=(18, 6))  # aumentei a largura para acomodar mais gráficos

def create_boxplot(feature):
    sns.boxplot(data=df[[feature]])
    plt.title(feature, fontsize=10)

features = ['hive temp', 'hive humidity', 'hive pressure',
            'weather temp', 'weather humidity', 'weather pressure']

for i, feature in enumerate(features):
    plt.subplot(1, 6, i+1)
    create_boxplot(feature)

plt.tight_layout()
plt.show()
```

#### (2) `Violin Plot`
-  ViolinPlot combina a visualização do boxplot juntamente com um plot de densidade (similar ao que vemos nos histogramas);
- Dentro do ViolinPlot encontramos os mesmos elementos que encontramos no BoxPlot;
- Mais detalhes: https://towardsdatascience.com/violin-plots-explained-fb1d115e023d
<span style="width:80%;text-align: center;"><img align = left src="https://miro.medium.com/v2/resize:fit:1040/format:webp/1*TTMOaNG1o4PgQd-e8LurMg.png" style="height:550px;"></span>

```python
plt.figure(figsize=(18, 3))  # largura ajustada para 6 gráficos lado a lado

def create_violinplot(feature):
    sns.violinplot(data=df[[feature]])
    plt.title(feature, fontsize=10)

features = ['hive temp', 'hive humidity', 'hive pressure',
            'weather temp', 'weather humidity', 'weather pressure']

for i, feature in enumerate(features):
    plt.subplot(1, 6, i+1)
    create_violinplot(feature)

plt.tight_layout()
plt.show()
```

#### (4) `Histplot`
* Frequência dos dados

```python
plt.figure(figsize=(18, 3))

def create_histogram(feature):
    sns.histplot(data = df[[feature]], kde=False)
    plt.title(feature, fontsize = 10)


features = ['hive temp', 'hive humidity', 'hive pressure',
            'weather temp', 'weather humidity', 'weather pressure']

for i, feature in enumerate(features):
    plt.subplot(1, 6, i+1)
    create_histogram(feature)

plt.tight_layout()

plt.show()


```
