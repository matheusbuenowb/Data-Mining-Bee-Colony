#Calculo da média da hive temp (temperatura interna da colmeia)

```python

df_limpo['hive temp'] = df[['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5']].mean(axis=1)
df_limpo.drop(columns=['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5'], inplace=True)
```

```python
df_limpo
```

```python
df_limpo.rename(columns={
    't': 'weather temp',
    'h': 'weather humidity',
    'p': 'weather pressure'
}, inplace=True)
```
