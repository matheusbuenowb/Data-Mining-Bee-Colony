#Pré-processamento e normalização de dados teste

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/arquivo_final.csv")
```

```python
df.info()
```

```python
missing_data = df.isnull().sum().to_frame().rename(columns={0: 'Numero total de faltantes'})
missing_data['% de faltantes'] = round((missing_data['Numero total de faltantes']/len(df))*100,2)
missing_data
```

```python
df_limpo = df.dropna()
```

```python
df_limpo.info()
```

```python
df_limpo
```

```python
df_limpo.to_csv('dadosTeste.csv', index=False)
```

```python
# Calcula a média das colunas t_i_1 até t_i_5
df_limpo['hive temp'] = df[['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5']].mean(axis=1)
df_limpo.drop(columns=['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5'], inplace=True)
```

```python
df_limpo
```

```python
df_limpo.rename(columns={
    't': 'weather temp',
    'h': 'weather humidity',
    'p': 'weather pressure'
}, inplace=True)
```

```python
df_limpo.to_csv('dadosTeste.csv', index=False)
```
